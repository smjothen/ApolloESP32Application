```python
import pandas as pd
import pandas_bokeh
pd.set_option('plotting.backend', 'pandas_bokeh')
pandas_bokeh.output_notebook()

from azure.cosmosdb.table.tableservice import TableService
from azure.cosmosdb.table.models import Entity

table_service = TableService(
    account_name='zapcloudprod',
    sas_token='st=2020-06-16T06%3A56%3A00Z&se=2022-06-19T06%3A56%3A00Z&sp=r&spk=ZAP000000&epk=ZAP000003&sv=2018-03-28&tn=observations&sig=fmHhaEQw77P%2FZNKfp5Pv8Q1XYPwy%2BZiE%2BjXWx08P8RE%3D'
)
task = table_service.get_entity('Observations', 'ZAP000001', '808_9005605823888721')                                                                                                      
if(task.value):
    print("connected")
else:
    raise RuntimeError("Failed to connect to table service")
    

    
```

```python
boot_messages = table_service.query_entities(
    'Observations',
    filter="PartitionKey eq 'ZAP000001' and RowKey lt '809_' and RowKey ge '808_' and value lt '[c' and value ge '[boot'",
    num_results=200
)

boot_frame = pd.DataFrame(boot_messages)


```

```python
key = 'uptime_ms: '

def get_connection_time(line):
    start_of_number = line.index(key)+len(key)
    return int(line[start_of_number:])
    

boot_frame['connection_time'] = boot_frame.apply(lambda row:get_connection_time(row['value']), axis=1)
boot_frame['connection_time'].plot_bokeh(kind='hist', bins=30)
```

```python
temperature_5_records = table_service.query_entities(
    'Observations',
    filter="PartitionKey eq 'ZAP000002' and RowKey lt '202_' and RowKey ge '201_'",
    num_results=999
)

temp5_frame = pd.DataFrame(temperature_5_records)
temp5_frame['value'].astype('float64')
clean_temp5_frame = pd.DataFrame()

#clean_temp5_frame['Timestamp'] = temp5_frame['Timestamp']
clean_temp5_frame['observedat'] = temp5_frame['observedat'].dt.tz_localize(None)
clean_temp5_frame['value'] = temp5_frame['value'].astype('float64')

temp5_frame = clean_temp5_frame
temp5_frame.set_index('observedat')

temp5_frame.plot_bokeh.line(x='observedat', y='value')
temp5_frame
```
